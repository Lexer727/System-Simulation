```python
def get_col(date):
    # if holiday or sunday return FT-{Month}
    if date in holidays.DE(subdiv='NW') or date.weekday() == 6:
        return "FT-" + date.strftime("%B")
    # if weekday return WT-{Month}
    if date.weekday() < 5:
        return "WT-" + date.strftime("%B")
    # default is saturday return SA-{Month}
    return "SA-" + date.strftime("%B")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ---- Map months to seasons ----
season_map = {
    'January': 'Winter', 'February': 'Winter',
    'March': 'Spring',  'April': 'Spring', 'May': 'Spring',
    'June': 'Summer',   'July': 'Summer',  'August': 'Summer',
    'September': 'Fall','October': 'Fall','November': 'Fall',
    'December': 'Winter'
}

seasons = ['Spring', 'Summer', 'Fall', 'Winter']
measurement_types = ['SA', 'FT', 'WT']

# ---- Build seasonal averages (96 rows for each season) ----
# result will be a dict: {mtype: DataFrame of shape (96, 4)}
seasonal_profiles = {}

for mtype in measurement_types:
    # Columns for this measurement type
    cols = [c for c in df.columns if c.startswith(mtype + '-')]
    
    # For each season, average across all month-columns in that season (row-wise)
    seasonal_avg = pd.DataFrame(index=df.index)
    for season in seasons:
        season_cols = [c for c in cols
                       if season_map[c.split('-')[1]] == season]
        # mean across those months, keep 96 rows
        seasonal_avg[season] = df[season_cols].mean(axis=1)
    
    seasonal_profiles[mtype] = seasonal_avg

# ---- Plot ----
for mtype, season_df in seasonal_profiles.items():
    plt.figure(figsize=(8,4))
    for season in seasons:
        plt.plot(season_df.index, season_df[season], label=season)
    tick_positions = np.arange(0, len(season_df.index), 4)
    plt.xticks(tick_positions, season_df.index[tick_positions], rotation=45)
    plt.title(f"{mtype} – Seasonal Average Profile")
    plt.xlabel("Quarter-hour Index (0–95)")
    plt.ylabel("Average Value")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
